```python
TABELA_OFICIAL = pd.merge(mesclado1, mesclado2, on='CO_VIA')
TABELA_OFICIAL = pd.merge(TABELA_OFICIAL, mesclado3, on='CO_URF')
TABELA_OFICIAL = pd.merge(TABELA_OFICIAL, mesclado4, on='CO_PAIS')
TABELA_OFICIAL = pd.merge(TABELA_OFICIAL, mesclado5, left_on='SG_UF_NCM', right_on='SG_UF')
TABELA_OFICIAL = pd.merge(TABELA_OFICIAL, mesclado6, on='CO_NCM')
```

```python
# Calcula o valor agregado
TABELA_OFICIAL['VL_AGREGADO'] = round(TABELA_OFICIAL['VL_FOB'] /TABELA_OFICIAL['KG_LIQUIDO'], 2)

# Filtro para valores maiores que 0.001.
TABELA_OFICIAL = TABELA_OFICIAL[
    (TABELA_OFICIAL['VL_FOB'] != 0) &
    (TABELA_OFICIAL['VL_FOB'] > 0.001) &
    (TABELA_OFICIAL['KG_LIQUIDO'] != 0) &
    (TABELA_OFICIAL['KG_LIQUIDO'] > 0.001)]

TABELA_OFICIAL['VL_AGREGADO'] = round(TABELA_OFICIAL['VL_FOB'] / TABELA_OFICIAL['KG_LIQUIDO'], 2)

# Remove VL_AGREGADO = 0
TABELA_OFICIAL = TABELA_OFICIAL[TABELA_OFICIAL['VL_AGREGADO'] != 0]
TABELA_OFICIAL = TABELA_OFICIAL.dropna()
print(TABELA_OFICIAL.head())
```

```python
#Calculo da FREQUÊNCIA em relação a movimentação dos produtos (NCM).
contagem_NCM = TABELA_OFICIAL['NO_NCM_POR'].value_counts()
total_de_ocorrencias_grupo = TABELA_OFICIAL.groupby(['CO_ANO', 'CO_MES', 'NO_NCM_POR']).size().reset_index(name='contagem_NCM_grupo')

contagem_NCM_df = contagem_NCM.reset_index(name='contagem_total_NCM').rename(columns={'index': 'NO_NCM_POR'})

df_merged = pd.merge(total_de_ocorrencias_grupo, contagem_NCM_df, on=['NO_NCM_POR'])

total_ocorrencias_mes_ano = TABELA_OFICIAL.groupby(['CO_ANO', 'CO_MES']).size().reset_index(name='total_ocorrencias_mes_ano')
df_merged = pd.merge(df_merged, total_ocorrencias_mes_ano, on=['CO_ANO', 'CO_MES'])
df_merged['FREQUENCIA'] = round((df_merged['contagem_NCM_grupo'] / df_merged['total_ocorrencias_mes_ano']) * 100, 2)

TABELA_OFICIAL = pd.merge(TABELA_OFICIAL, df_merged[['CO_ANO', 'CO_MES', 'NO_NCM_POR', 'FREQUENCIA']],
                        on=['CO_ANO', 'CO_MES', 'NO_NCM_POR'], how='left')
print(df_merged.head())
```

```python
VIAS = ['AEREA', 'FERROVIARIA', 'MARITIMA', 'RODOVIARIA']
TABELA_OFICIAL = TABELA_OFICIAL[TABELA_OFICIAL['NO_VIA'].isin(VIAS)]
```

```python
# Exemplo de padronização
def padronizar_coluna(df, col):
    if col in df.columns:
        # Converte para string, remove espaços extras, transforma em maiúsculas
        df[col] = df[col].astype(str).str.strip().str.upper()
    return df

TABELA_OFICIAL = padronizar_coluna(TABELA_OFICIAL, 'NO_UF')
TABELA_OFICIAL = padronizar_coluna(TABELA_OFICIAL, 'NO_URF') # Se TABELA_OFICIAL também tiver URF/PAIS
TABELA_OFICIAL = padronizar_coluna(TABELA_OFICIAL, 'NO_PAIS')

mesclado8 = padronizar_coluna(mesclado8, 'NO_UF')
mesclado9 = padronizar_coluna(mesclado9, 'NO_URF')
mesclado10 = padronizar_coluna(mesclado10, 'NO_PAIS')

# Se a coluna de junção em mesclado7 for 'NO_UF' também, padronize mesclado7['NO_UF']
```
