```python
print(mesclado1.columns)
print(mesclado2.columns)
print(mesclado3.columns)
print(mesclado4.columns)
print(mesclado5.columns)
print(mesclado6.columns)
```

```python
tabela_final = pd.merge(mesclado1, mesclado2, on='CO_VIA')
tabela_final = pd.merge(tabela_final, mesclado3, on='CO_URF')
tabela_final = pd.merge(tabela_final, mesclado4, on='CO_PAIS')
tabela_final = pd.merge(tabela_final, mesclado5, left_on='SG_UF_NCM', right_on='SG_UF')
tabela_final = pd.merge(tabela_final, mesclado6, on='CO_NCM')
```

```python
# Calcula o valor agregado
tabela_final['VL_AGREGADO'] = round(tabela_final['VL_FOB'] / tabela_final['KG_LIQUIDO'], 2)

# Filtro para valores maiores que 0.001.
tabela_final = tabela_final[
    (tabela_final['VL_FOB'] != 0) &
    (tabela_final['VL_FOB'] > 0.001) &
    (tabela_final['KG_LIQUIDO'] != 0) &
    (tabela_final['KG_LIQUIDO'] > 0.001)]

tabela_final['VL_AGREGADO'] = round(tabela_final['VL_FOB'] / tabela_final['KG_LIQUIDO'], 2)

# Remove VL_AGREGADO = 0
tabela_final = tabela_final[tabela_final['VL_AGREGADO'] != 0]
tabela_final = tabela_final.dropna()
print(tabela_final.head())
```

```python
#Calculo da FREQUÊNCIA em relação a movimentação dos produtos (NCM).
contagem_NCM = tabela_final['NO_NCM_POR'].value_counts()
total_de_ocorrencias_grupo = tabela_final.groupby(['CO_ANO', 'CO_MES', 'NO_NCM_POR']).size().reset_index(name='contagem_NCM_grupo')

contagem_NCM_df = contagem_NCM.reset_index(name='contagem_total_NCM').rename(columns={'index': 'NO_NCM_POR'})

df_merged = pd.merge(total_de_ocorrencias_grupo, contagem_NCM_df, on=['NO_NCM_POR'])

total_ocorrencias_mes_ano = tabela_final.groupby(['CO_ANO', 'CO_MES']).size().reset_index(name='total_ocorrencias_mes_ano')
df_merged = pd.merge(df_merged, total_ocorrencias_mes_ano, on=['CO_ANO', 'CO_MES'])
df_merged['FREQUENCIA'] = round((df_merged['contagem_NCM_grupo'] / df_merged['total_ocorrencias_mes_ano']) * 100, 2)

tabela_final = pd.merge(tabela_final, df_merged[['CO_ANO', 'CO_MES', 'NO_NCM_POR', 'FREQUENCIA']],
                        on=['CO_ANO', 'CO_MES', 'NO_NCM_POR'], how='left')
print(df_merged.head())
```

```python
VIAS = ['AEREA', 'FERROVIARIA', 'MARITIMA', 'RODOVIARIA']
tabela_final = tabela_final[tabela_final['NO_VIA'].isin(VIAS)]
```
